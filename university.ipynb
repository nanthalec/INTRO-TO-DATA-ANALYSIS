```python
#how to find the different ones inside of Type 
university['Type'].unique()
```

```python
# 1 information for 4-year, Private not-for-profit
private_noprofit = university[university['Type'] == '4-year, Private not-for-profit']
private_noprofit =private_noprofit.describe()

#rename
private_noprofit= private_noprofit.rename(columns={"Size": "private_noprofit"})
private_noprofit
```

```python
# 2 information for 4-year, primarily associate's, Public
primary_public = university[university['Type'] == "4-year, primarily associate's, Public"]
primary_public = primary_public.describe()

#rename
primary_public = primary_public.rename(columns={"Size": "primary_public"})
primary_public
```

```python
# 3 4-year, primarily associate's, Private for-profit
primarily_private_profit = university[university['Type'] =="4-year, primarily associate's, Private for-profit"]
primarily_private_profit= primarily_private_profit.describe()

#rename
primarily_private_profit = primarily_private_profit.rename(columns={"Size": "primarily_private_profit"})
primarily_private_profit
```

```python
# 4 information for 4-year, Private for-profit
private_profit = university[university['Type'] =='4-year, Private for-profit']
private_profit= private_profit.describe()

# rename
private_profit = private_profit.rename(columns={"Size": "private_profit"})
private_profit
```

```python
# 5 information for  4-year, Public 
public = university[university['Type'] =='4-year, Public']
public= public.describe()

#rename 
public = public.rename(columns={"Size": "public"})
public
```

```python
# 6 information for "4-year, primarily associate's, Private not-for-profit"
primarily_private_nonprofit = university[university['Type'] == "4-year, primarily associate's, Private not-for-profit"]
primarily_private_nonprofit = primarily_private_nonprofit.describe()

#rename 
primarily_private_nonprofit = primarily_private_nonprofit.rename(columns={"Size": "primarily_private_nonprofit"})
primarily_private_nonprofit
```

```python
#how to combine all info type into one database
one=pd.merge(
    private_noprofit, 
    primary_public,
    right_index=True, left_index=True
)
one
```

```python
two=pd.merge(
    primarily_private_profit, 
    private_profit,
    right_index=True, left_index=True
)
two
```

```python
three=pd.merge(
    public, 
    primarily_private_nonprofit,
    right_index=True, left_index=True
)
three
```

```python
one_more=pd.merge(
    one, 
    two,
    right_index=True, left_index=True
)
one_more
```

```python
final=pd.merge(
    one_more,
    three,
    right_index=True, left_index=True
)
final
```

```python
university[['address2','state']] = university['address_'].str.rsplit(', ',1, expand=True)
```

```python
university.drop('address_', axis=1, inplace=True)
```

```python
university[['address3','city']] = university['address2'].str.rsplit(', ',1, expand=True)
university
```

```python
university.drop('address2', axis=1, inplace=True)
university
```

```python
#rearange the columns
university= university[['name','Website','Type','Size','address3','city','state','zip']]
university
```

```python
# Private or Public
# for-profit or not-for-profit
# primarily associates
```

```python
university['Public/Private'] = np.where(university['Type'].str.contains("Private")>0, 'Private', 'Public')
```

```python
university['Public/Private'].value_counts()
```

```python
university
```

```python
# for-profit or not-for-profit
```

```python
university['For-Profit/NFP'] = np.where(university['Type'].str.contains(" for")>0, 'For-Profit', 'NFP')
```

```python
university
```

```python
# primarily associates
```

```python
university['Primarily_Associates?'] = np.where(university.Type.str.contains('prim')>0, 'Yes', 'No')
```

```python
university[university['Type'].str.contains("prim")>0]
```

```python
# Using the new columns youve made above (and excluding the original type column), 
# find both the mean and median size of each university by State, by Private or Public, 
```

```python
university.groupby(['state', 'Public/Private', 'Primarily_Associates?']).aggregate([np.mean, np.median])
```

```python
# Using the new columns youve made above (and excluding the original type column), 
# and by whether the university is primarily associate's or not.
```

```python
university.groupby(['state', 'Public/Private', 'Primarily_Associates?']).aggregate(np.sum)
```
